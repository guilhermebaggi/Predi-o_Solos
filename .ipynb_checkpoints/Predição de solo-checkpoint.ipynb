Pré-Processamento

```python
def snv(X):
  
    X_mean = np.mean(X, axis=1, keepdims=True)
    X_std = np.std(X, axis=1, keepdims=True)

    X_std[X_std == 0] = 1e-9

    X_snv = (X - X_mean) / X_std
    return X_snv
```

```python
X_snv = snv(X)
```

```python
def msc(X, reference=None):
    
    if reference is None:
        reference = np.mean(X, axis=0)

    X_msc = np.zeros_like(X)

    for i in range(X.shape[0]):
        x = X[i, :]
        A = np.vstack([reference, np.ones(len(reference))]).T
        slope, intercept = np.linalg.lstsq(A, x, rcond=None)[0]

        X_msc[i, :] = (x - intercept) / slope

    return X_msc
```

```python
X_msc = msc(X)
```

```python
def avaliarPLS(X, y):
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

    pipe = Pipeline([
        ('pls', PLSRegression())
    ])

    parameters = {
        'pls__n_components': range(1, min(40, X.shape[1]) + 1)

    }

    cv=KFold(n_splits=10, shuffle=True, random_state=42)
    model = GridSearchCV(pipe, param_grid=parameters, scoring='neg_mean_squared_error', n_jobs=-1, cv=cv)
    model.fit(X_train, y_train)

    print("Melhores parâmetros do modelo:", model.best_params_)

    y_val = model.predict(X_test)
    validation_r2 = r2_score(y_test, y_val)
    validation_rmse = np.sqrt(mean_squared_error(y_test, y_val))
    validation_mae = mean_absolute_error(y_test, y_val)
    validation_sd = np.std(y_test)
    validation_rpd = validation_sd / validation_rmse

    print("\nMÉTRICAS DA VALIDAÇÃO")
    dados_val = {
        "Metric": ["RMSE", "R²", "RPD", "MAE"],
        "Value": [validation_rmse, validation_r2, validation_rpd, validation_mae],
    }
    df_val = pd.DataFrame(dados_val)
    print(df_val)

    y_cal = model.predict(X_train)
    calibration_r2 = r2_score(y_train, y_cal)
    calibration_rmse = np.sqrt(mean_squared_error(y_train, y_cal))
    calibration_mae = mean_absolute_error(y_train, y_cal)
    calibration_sd = np.std(y_train)
    calibration_rpd = calibration_sd / calibration_rmse

    print("\nMÉTRICAS DA CALIBRAÇÃO")
    dados_cal = {
        "Metric": ["RMSE", "R²", "RPD", "MAE"],
        "Value": [calibration_rmse, calibration_r2, calibration_rpd, calibration_mae],
    }
    df_cal = pd.DataFrame(dados_cal)
    print(df_cal)
```

```python
def savgol(X):
    
    w = 17
    p = 2
    d = 2
    
    sgX = savgol_filter(X, w, polyorder = p, deriv = d)
    sgX.shape

    fig, axs = plt.subplots(2, 1)  
    axs[0].plot(wl, X.T)  
    axs[0].set_xlabel('wavenumber')  
    axs[0].set_ylabel('Absorbancia')
    axs[0].grid(True)  
    axs[0].set_title("Sinal Original") 
    axs[1].plot(wl, sgX.T)
    axs[1].set_xlabel('wavenumber')
    axs[1].set_title("Sinal suavisado com Savitzky-Golay")
    axs[1].grid(True)
    fig.tight_layout()  
    plt.show()

    return sgX
```

```python
def PCA_otimizado(X, y, max_components=40, test_size=0.3, random_state=42):
    
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=test_size, random_state=random_state)
    
    pipe = Pipeline([
        ('scaler', StandardScaler()),
        ('pca', PCA()),
        ('pls', PLSRegression())
    ])

    param_grid = {
        'pca__n_components': range(1, 163),
        'pls__n_components': range(1, 162)
    }

    cv = KFold(n_splits=10, shuffle=True, random_state=random_state)
    grid = GridSearchCV(pipe, param_grid, cv=cv, scoring='r2', n_jobs=-1, error_score=np.nan)    
    grid.fit(X_train, y_train)

    best_model = grid.best_estimator_
    best_params = grid.best_params_
    print("Melhor n_components PCA:", best_params['pca__n_components'])
    print("Melhor n_components PLS:", best_params['pls__n_components'])
    print("Melhor score (CV treino R²):", grid.best_score_)

    return best_model, best_params
```

```python
def aplicarPCA(X, n_components):
    
    scaler = StandardScaler()
    X_scaled = scaler.fit_transform(X)
    
    pca = PCA(n_components= n_components)
    X_pca = pca.fit_transform(X_scaled)
    
    #print("shape dos dados apos o pca: ",X_pca.shape)
    #print("Variância explicada por componente (pca):", pca.explained_variance_ratio_)
    #print("Variância acumulada (pca):", np.cumsum(pca.explained_variance_ratio_))
    #print("Valores singulares (pca):", pca.singular_values_ )
    
    plt.figure(figsize=(10, 5))
    plt.bar(range(1, len(pca.explained_variance_ratio_) + 1), pca.explained_variance_ratio_, alpha=0.7, align='center')
    plt.step(range(1, len(pca.explained_variance_ratio_) + 1), np.cumsum(pca.explained_variance_ratio_), where='mid', linestyle='--', color='red')
    plt.xlabel('Componente Principal')
    plt.ylabel('Variância Explicada')
    plt.title('Variância Explicada por PCA')
    plt.grid(True)
    plt.show()
    
    return X_pca
    
```

```python
def PCR_otimizado(X, y, test_size=0.3, random_state=42):
    X_train, X_test, y_train, y_test = train_test_split(
        X, y, test_size=test_size, random_state=random_state
    )

    pipe = Pipeline([
        ('pca', PCA()),
        ('reg', LinearRegression())
    ])

    param_grid = {
        'pca__n_components': range(1, min(40, X.shape[1]) + 1)
    }

    cv = KFold(n_splits=10, shuffle=True, random_state=random_state)
    model = GridSearchCV(
        pipe, param_grid=param_grid,
        scoring='neg_mean_squared_error',
        n_jobs=-1, cv=cv
    )
    model.fit(X_train, y_train)

    print("Melhor número de componentes:", model.best_params_['pca__n_components'])

    y_val = model.predict(X_test)
    validation_r2 = r2_score(y_test, y_val)
    validation_rmse = np.sqrt(mean_squared_error(y_test, y_val))
    validation_mae = mean_absolute_error(y_test, y_val)
    validation_sd = np.std(y_test)
    validation_rpd = validation_sd / validation_rmse

    print("\nMÉTRICAS DA VALIDAÇÃO")
    dados_val = {
        "Metric": ["RMSE", "R²", "RPD", "MAE"],
        "Value": [validation_rmse, validation_r2, validation_rpd, validation_mae],
    }
    df_val = pd.DataFrame(dados_val)
    print(df_val)

    y_cal = model.predict(X_train)
    calibration_r2 = r2_score(y_train, y_cal)
    calibration_rmse = np.sqrt(mean_squared_error(y_train, y_cal))
    calibration_mae = mean_absolute_error(y_train, y_cal)
    calibration_sd = np.std(y_train)
    calibration_rpd = calibration_sd / calibration_rmse

    print("\nMÉTRICAS DA CALIBRAÇÃO")
    dados_cal = {
        "Metric": ["RMSE", "R²", "RPD", "MAE"],
        "Value": [calibration_rmse, calibration_r2, calibration_rpd, calibration_mae],
    }
    df_cal = pd.DataFrame(dados_cal)
    print(df_cal)

    return model.best_estimator_, model.best_params_['pca__n_components'], y_val
```

Remoção de outliers

PCR + Distancia de Mahalanobis

```python
def PCR_mahalanobis_grid(X, y, max_components=40, alpha=0.01):
    n_temp = min(max_components, X.shape[1])
    pca_temp = PCA(n_components=n_temp)
    X_scores = pca_temp.fit_transform(X)

    cov = np.cov(X_scores.T)
    inv_cov = np.linalg.inv(cov)
    mean_vec = np.mean(X_scores, axis=0)
    d_mahal = np.array([mahalanobis(x, mean_vec, inv_cov) for x in X_scores])

    threshold = np.sqrt(chi2.ppf(1 - alpha, df=n_temp))
    inliers = d_mahal < threshold
    X_clean = X[inliers]
    y_clean = y[inliers]

    pipeline = Pipeline([
        ('pca', PCA()),
        ('reg', LinearRegression())
    ])

    param_grid = {
        'pca__n_components': list(range(1, min(X_clean.shape[1], max_components) + 1))
    }

    grid = GridSearchCV(pipeline, param_grid, cv=10, scoring='r2')
    grid.fit(X_clean, y_clean)

    best_model = grid.best_estimator_
    best_n_components = grid.best_params_['pca__n_components']
    y_pred = best_model.predict(X_clean)

    mae = mean_absolute_error(y_clean, y_pred)
    mse = mean_squared_error(y_clean, y_pred)
    rmse = np.sqrt(mse)
    r2 = best_model.score(X_clean, y_clean)


    print(f"Melhor numero de componente:{best_n_components}")
    print(f"MAE:{mae:.2f}")
    print(f"MSE:{mse:.2f}")
    print(f"RMSE:{rmse:.3f}")
    print(f"R²:{r2:.2f}")
    print(f"n_amostras:{len(y_clean)}")
    print(f"n_outliers:{len(y) - len(y_clean)}")
```

```python
PCR_mahalanobis_grid(X, yNitrogen, max_components=40, alpha=0.01)
PCR_mahalanobis_grid(X, yCarbon, max_components=40, alpha=0.01)
```

```python
PCR_mahalanobis_grid(X_sg, yNitrogen, max_components=40, alpha=0.01)
PCR_mahalanobis_grid(X_sg, yCarbon, max_components=40, alpha=0.01)
```

LOF

```python
def remover_outliers_lof(X, y):
    lof = LocalOutlierFactor(n_neighbors=20, contamination=0.1)
    y_pred = lof.fit_predict(X)
    mask = y_pred != -1
    X_LOF, y_LOF = X[mask], y[mask]
    print(f"LOF: {X_LOF.shape[0]} amostras mantidas de {X.shape[0]}")
    return X_LOF, y_LOF
    
```

DBSCAN

```python
def remover_outliers_dbscan(X, y):
    db = DBSCAN(eps=0.5, min_samples=5)
    labels = db.fit_predict(X)
    mask = labels != -1
    X_DBSCAN, y_DBSCAN = X[mask], y[mask]
    print(f"DBSCAN: {X_DBSCAN.shape[0]} amostras mantidas de {X.shape[0]}")
    return X_DBSCAN, y_DBSCAN
```

ISOLATION FOREST

```python
def remover_outliers_isolation_forest(X, y):
    iso = IsolationForest(contamination='auto', random_state=42)
    y_pred = iso.fit_predict(X)
    mask = y_pred != -1
    X_IF, y_IF = X[mask], y[mask]
    print(f"Isolation Forest: {X_IF.shape[0]} amostras mantidas de {X.shape[0]}")
    return X_IF, y_IF
```
